**Write a Python program to find the average CGPA of your class in a dictionary**

```python
dictionary = {'Sarwar':2.8,'Fatima':3.5,'Sadia':3.9,'Deepak':3.3,'Hammad':3.7}
values = 0
for i in dictionary.values():
    values += i
length = len(dictionary)
average = values/length
print("Average CGPA of the class =",average)
```
