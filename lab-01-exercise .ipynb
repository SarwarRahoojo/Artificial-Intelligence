Write a function that takes input a number, which is temperature in degree Celsius and return the
temperature in kelvin and Fahrenheit

```python
def tepm(temperature):
    temp_kelvin = float(temperature)+273.15
    print("Temperature in Kelvin is ",temp_kelvin)
    temp_fahrenheit = float(temperature)*(9/5)+32
    print("Temperature in Fahrenheit is ",temp_fahrenheit )
tmp = input ("Enter temperature in degree Celsius")
tepm(tmp)
```

Create a list of even numbers and odd numbers with the help of for loop and append function.
Than combine both lists together.

```python
L1 = []
L2 = []
L3 = []
num = input ("Print Odd and Even Numbers Until = ")
num = int(num)
for i in range(num):
    if(i%2==0):
        L1.append(i)
        print(L1)
    elif(i%2==1):
        L2.append(i)
        print(L2)
L3.extend(L1)
L3.extend(L2)
L3.sort()
L3
```
